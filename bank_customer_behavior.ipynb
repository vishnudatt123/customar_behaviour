```python
import matplotlib.pyplot as plt

# How many people are in which education level
df['education'].value_counts().plot(kind='bar')
plt.title("Education Distribution")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.show()
```

```python
# This graph shows how income is distributed among customers.
plt.hist(df['income'])
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# This graph shows how many customers have taken loans. 
df['loan'].value_counts().plot(kind='bar')
plt.title("Loan Distribution")
plt.xlabel("Loan (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# What % of people take loans
df['loan'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Loan Distribution")
plt.ylabel("")
plt.show()
```

```python
# This plot shows the relationship between working hours and income.
plt.scatter(df['working_hours'], df['income'])
plt.show()
```
